```python
def fourierparpartie(data, debut, fin, taillefenetre):
    print("Debut fourier ..................")
    res = []
    reste = len(data) % taillefenetre
#     newtaille = int((len(data) - reste) / taillefenetre)
    for k in range(0, (fin-debut-1)):
        liste = data[taillefenetre * k:taillefenetre*(k+1)]
        tmp = numpy.fft.fft2(liste)
        res = res + [tmp]
    liste = data[(len(data) - reste): len(data)]
    res = res + [numpy.fft.fft2(liste)]
    return res

a = fourierparpartie(data, 0, 15, 10)
```
